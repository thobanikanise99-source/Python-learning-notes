# Types of Python Collections

| Collection | Ordered | changeable | Allows Duplicates |
|---|---|---|---|
| List | Yes | Yes | Yes |
| Tuple | Yes | No | Yes |
| Set | No | Yes | No |
| Dictionary | Yes | Yes | Keys Cannot Duplicate |
