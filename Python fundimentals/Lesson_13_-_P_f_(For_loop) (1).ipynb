## For Loop (Iteration over a sequence)

A `for` loop goes through each item in a sequence one by one.

**Flow:**
- Start from first item
- Execute code
- Move to next item
- Stop when sequence ends




| Term | Meaning|
|-|-|
|Loop|Repeating code|
|iteration| one repititon|
|collection| group of items
|loop variable|temporary variable used in loopd

```python
products = ["Bread", "Milk", "Eggs", "Bread", "Juice"]
```

```python
for item in products:
    print(item)
```

```python
# iteration 1
item = products[0]
print(item)
```

```python
# iteration 2
item = products[1]
print(item)
```

## Loop with if statement

```python
balances = [10, 2, 30, 1]

for balance in balances:

    if balance < 5:
        print(balance)
```

## Loop through Dictionary

```python
customers = {
    "Alice": 10,
    "Bob": 2,
    "David": 20
}
```

```python
#loop through keys
for customer in customers:
    print(customer)
```

```python
#loop through values
for quantity in customers.values():
    print(quantity)
```

```python
print(customers.items())
```

```python
#loop through both key and value

for customer, quantity in customers.items():
    print(customer, quantity)
```

## Loop with a counter

```python
balances = [10, 2, 30, 1]

counter = 0

for counter in range(len(balances)):

    if balances[counter] < 5:
        print(balances[counter])
    
    counter = counter + 1
```
