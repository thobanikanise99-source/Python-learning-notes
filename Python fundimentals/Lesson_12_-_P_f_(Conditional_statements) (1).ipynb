
# What are Conditional Statements?
Conditional statements allow a program to make decisions.

| Operator | Meaning               |
| -------- | --------------------- |
| `==`     | Equal to              |
| `!=`     | Not equal to          |
| `>`      | Greater than          |
| `<`      | Less than             |
| `>=`     | Greater than or equal |
| `<=`     | Less than or equal    |

# if...else

```python
marks = 45

if marks >= 50:
    print("Pass")
else:
    print("Fail")
```


# if...elif...else

Python checks from top to bottom

1. is the score >= 80
2. if not, is score >=60
3. if not, is score >= 50
4. Else -> Fail

```python
score = 75


if score >= 80: #ranges 80 - 100
    print("Distinction")
elif score >= 60: #ranges 60 - 80
    print("Merit")
elif score >= 50:#ranges 50 - 60
    print("Pass")
else:#ranges 0 - 50
    print("Fail")
```

# Logical Operators

| Operator | Meaning                      |
| -------- | ---------------------------- |
| `and`    | Both conditions must be true |
| `or`     | One condition must be true   |
| `not`    | Reverses condition           |
