Python collections are used to store multiple values in a single variable.

Collections are important in Data Analytics because they help us work with:
- customer records
- sales data
- product information
- KPIs
- datasets

```python
products = ["Laptop", "Phone", "Tablet"]

products = list(("Laptop", "Phone", "Tablet"))

mixed_list = ["Apple", 200, True]
```

```python
customers = ["John", "Sarah", "Mike"]

print(customers[0])
print(customers[2])
```

## Change item

```python
customers[0] = "Thabo"

print(customers)
```

## append()

```python
customers.append("David")
print(customers)
```

## insert()

```python
customers.insert(1, "Emma")
print(customers)
```

## remove()

```python
customers.remove("Mike")
print(customers)
```

## pop() -- last in first out

```python
customers.pop()
print(customers)
```

sort()

```python
customers.sort()
print(customers)
```

# Set

- Unorder
- Changeable
- no duplicates

```python
regions = {"Gauteng", "Limpopo", "Gauteng"}

print(regions)
```

```python
regions = {"Gauteng", "Limpopo", "Gauteng"}
regions = set(("Gauteng", "Limpopo", "Gauteng"))
```

```python
regions.add("Western Cape")
print(regions)
```

```python
reg2 = {"Free State", "Eastern Cape"}

print(reg2)
```

```python
final_reg = regions.union(reg2)

print(final_reg)
```

# Final Recap
Lists

Use when:

Tuples
Use when:

Sets
